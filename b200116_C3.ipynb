```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.naive_bayes import BernoulliNB
from sklearn.metrics import classification_report,accuracy_score,confusion_matrix
from sklearn.tree import DecisionTreeClassifier
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree
```

```python
df=pd.read_csv("train_titanic.csv")
df
```

```python
df['Age']=df['Age'].fillna(df['Age']).mean()

df.isnull().sum()
```

```python
df['Sex']=pd.Categorical(df['Sex']).codes
df['Embarked']=pd.Categorical(df['Embarked']).codes
col=['Sex','Age','Embarked']
X=df[col]
y=df['Survived']
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25)
```

```python



#LINEAR REGRESSION


```

```python
model1=LinearRegression()
model1.fit(X_train,y_train)
```

```python
y_predict=model1.predict(X_test)
y_predict= [1 if pred >= 0.5 else 0 for pred in y_predict]
```

```python
print("Accuracy : ",accuracy_score(y_test,y_predict))
print("Classification Report : ",classification_report(y_test,y_predict))
print("Confusion Matrix : ",confusion_matrix(y_test,y_predict))
```

```python




#lOGISTIC REGRESSION





```

```python
model2=LogisticRegression()
model2.fit(X_train,y_train)
```

```python
y_predict1=model2.predict(X_test)
```

```python
print("Accuracy : ",accuracy_score(y_test,y_predict1))
print("Classification Report : ",classification_report(y_test,y_predict1))
print("Confusion Matrix : ",confusion_matrix(y_test,y_predict1))
```

```python

#SVM..............................





model3=SVC()
model3.fit(X_train,y_train)
```

```python
y_predict2=model3.predict(X_test)
print("Accuracy : ",accuracy_score(y_test,y_predict2))
print("Classification Report : ",classification_report(y_test,y_predict2))
print("Confusion Matrix : ",confusion_matrix(y_test,y_predict2))
```

```python
#DECISION TREE.......................





```

```python
model=DecisionTreeClassifier(max_depth=3)
```

```python
model.fit(X_train,y_train)
```

```python
plt.figure(figsize=(20, 10))
plot_tree(
    model, 
    feature_names=X_train.columns, 
    class_names=model.classes_.astype(str),
    filled=True, 
    rounded=True
)
plt.show()
```
